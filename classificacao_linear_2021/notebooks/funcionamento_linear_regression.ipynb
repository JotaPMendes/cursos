**DEFININDO A EQUAÇÃO DA RETA NO PYPLOT**

```python
import numpy as np
import matplotlib.pyplot as plt

a = -1
b = 4
c = 0.4

# ax + by + c = 0
# y = (-a*x - c) / b

def plotline (a, b, c):
  x = np.linspace (-2, 4, 50)    # linspace(intervalo1, intervalo2, nº de valores)
  y = (-a*x - c)/b

  plt.axvline(0, -1, 1, color='k', linewidth=1)  # plt.axvline: cria o eixo vertical do plano
  plt.axhline(0, -2, 4, color='k', linewidth=1)  # plt.axvline: cria o eixo horizontal do plano
  plt.plot(x, y)    # plt.plot: irá plotar o gráfico
  plt.grid(True)    # plt.grid: colabora com a visualização

plotline(a, b, c)
```

**INSERINDO PONTOS NO PLANO**

```python
p1 = (2, 0.4)
p2 = (1, 0.6)
p3 = (3, -0.4)


plotline(a, b, c)
plt.plot(p1[0], p1[1], color = 'b', marker = 'o')
plt.plot(p2[0], p2[1], color = 'r', marker = 'o')
plt.plot(p3[0], p3[1], color = 'g', marker = 'o')
```

```python
ret1 = a*p1[0] + b*p1[1] + c
print("%.2f" % ret1)

ret2 = a*p2[0] + b*p2[1] + c
print("%.2f" % ret2)

ret3 = a*p3[0] + b*p3[1] + c
print("%.2f" % ret3)
```

A equação da reta para o perceptron de classificação linear, ao invés da tradicional "ax + bx + c = y", será "w1x1 + w2x2 + b = y". Sendo x, o número de planos e w, os pesos da rede. Para dimensões > 2, os modelos passam a se chamar de hiperplanos, compostos por:


*   Um peso (w)i para cada dimensão (x)i
*   Um único viés "b"

```python
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt

np.random.seed(46)

X, Y = make_classification(n_features=2, n_redundant=0, n_informative=1,
                             n_clusters_per_class=1)

plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y,
            edgecolor='k')

p = X[10]
print(Y[10])
plt.plot(p[0], p[1], marker='^', markersize=20)
```

```python
def plotmodel (w1, w2, b):

  plt.scatter(X[:, 0], X[:, 1], marker='o', c=Y,
            edgecolor='k')

  xmin, xmax = plt.gca().get_xlim()
  ymin, ymax = plt.gca().get_ylim()

  x = np.linspace (-2, 4, 50)    # linspace(intervalo1, intervalo2, nº de valores)
  y = (-w1*x - b)/w2

  plt.axvline(0, -1, 1, color='k', linewidth=1)  # plt.axvline: cria o eixo vertical do plano
  plt.axhline(0, -2, 4, color='k', linewidth=1)  # plt.axvline: cria o eixo horizontal do plano
  plt.plot(x, y)    # plt.plot: irá plotar o gráfico
  plt.grid(True)    # plt.grid: colabora com a visualização

  plt.xlim(xmin, xmax)
  plt.ylim(ymin, ymax)

w1 = 5
w2 = 1
b = 0.4

plotmodel(w1, w2, b)

p = (-1, 1)
print(w1 * p[0] + w2 * p[1] + b)
```

```python
def classify (ponto, w1, w2, b):
  ret = w1 * ponto[0] + w2 * ponto[1] + b

  if ret >= 0:
    return 1, 'yellow'
  else:
    return 0, 'blue'

p = (2, -1)
classe, cor = classify(p, w1, w2, b)
print(classe, cor)

plotmodel(w1, w2, b)
plt.plot(p[0], p[1], marker='^', color = cor, markersize = 30)
```

```python
acertos = 0
for k in range(len(X)):
  categ, _ = classify(X[k], w1, w2, b)
  if categ == Y[k]:
    acertos +=1

print("Acurácia: {0}" .format(acertos/len(X)))
```
