# Pétalas e sépalas de uma flor

```python
df = pd.read_csv("/content/curso-matplotlib-preparando_ambiente/iris.csv")
```

```python
df.head()
```

```python
fig = plt.figure(figsize = (10,5))
eixo = fig.add_axes([0,0,1,1])

cores = {'Iris-setosa' : 'r', "Iris-versicolor" : 'b', "Iris-virginica" : 'g'}
marcadores = {'Iris-setosa' : 'x', "Iris-versicolor" : 'o', "Iris-virginica" : 'v'}

for especie in df['espécie'].unique():
  tmp = df[df['espécie'] == especie]
  eixo.scatter(tmp['comprimento_sépala'], tmp['largura_sépala'],
               color = cores[especie], marker = marcadores[especie],
               s = 40)

eixo.set_title("Gráfico de dispersão", fontsize = 20, pad = 15)
eixo.set_xlabel("Comprimento da sépala", fontsize = 10)
eixo.set_ylabel("Largura da sépala", fontsize = 10)
eixo.tick_params(labelsize = 10)
eixo.legend(cores, fontsize = 10)
```

```python
fig = plt.figure(figsize = (7,4))
eixo = fig.add_axes([0,0,1,1])

eixo.boxplot(df['largura_pétala'])
eixo.set_title("Gráfico de caixa", fontsize = 15, pad = 10)
eixo.set_xticklabels(['largura_pétala'])
```

```python
fig = plt.figure(figsize = (7,4))
eixo = fig.add_axes([0,0,1,1])

cores = ['red', 'blue', 'orange', 'green']

caixas = eixo.boxplot(df.drop('espécie', axis = 1).values, patch_artist = True)
eixo.set_title("Gráfico de caixa", fontsize = 15, pad = 10)
eixo.set_xticklabels(df.drop('espécie', axis = 1).columns)

for caixa, cor in zip(caixas['boxes'], cores):
  caixa.set(color = cor)

for outlier in caixas['fliers']:
  outlier.set(marker = 'x', markersize = 8)
```

```python
fig = plt.figure(figsize = (7,4))
eixo = fig.add_axes([0,0,1,1])

eixo.hist(df['comprimento_pétala'], bins = 20, density = True)
eixo.set_title('Histograma', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da pétala', fontsize = 15)
eixo.grid(True)
```

```python
fig = plt.figure(figsize = (7,4))
eixo = fig.add_axes([0,0,1,1])

media, desvio = df['comprimento_pétala'].mean(), df['comprimento_pétala'].std()

eixo.hist(df['comprimento_pétala'], bins = 20)
eixo.set_title('Histograma', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da pétala', fontsize = 15)
eixo.grid(True)

eixo.annotate('$media = {0:.2f}$\n$desvio = {1:.2f}$' .format(media, desvio),
              xy = (4.5,20), fontsize = 15)
eixo.axvline(media, color = 'black', linestyle = '--')
eixo.annotate('média', xy = (media - 1, 28), fontsize = 15)
eixo.axvline(df['comprimento_pétala'].median(), color = 'g', linestyle = '--')
eixo.annotate('mediana', xy = (df['comprimento_pétala'].median() + 0.1, 31), fontsize = 15)
```

```python
fig = plt.figure(figsize = (7,4))
eixo = fig.add_axes([0,0,1,1])

df_iv = df[df['espécie'] == 'Iris-versicolor']

media, desvio = df_iv['comprimento_pétala'].mean(), df_iv['comprimento_pétala'].std()

eixo.hist(df_iv['comprimento_pétala'], bins = 20)
eixo.set_title('Iris Versicolor', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da pétala', fontsize = 15)
eixo.grid(True)

eixo.annotate('$media = {0:.2f}$\n$desvio = {1:.2f}$' .format(media, desvio),
              xy = (4.6, 6), fontsize = 15)
eixo.axvline(media, color = 'black', linestyle = '--')
eixo.annotate('média', xy = (media - 0.3, 5.5), fontsize = 15)
eixo.axvline(df_iv['comprimento_pétala'].median(), color = 'g', linestyle = '--')
eixo.annotate('mediana', xy = (df_iv['comprimento_pétala'].median() -0.5, 6.5), fontsize = 15,
              color = 'g')

fig.savefig('histograma_iv.png', bbox_inches = 'tight')
```

```python
fig = plt.figure(figsize = (7,4))
eixo = fig.add_axes([0,0,1,1])

df_is = df[df['espécie'] == 'Iris-setosa']

media, desvio = df_is['comprimento_pétala'].mean(), df_is['comprimento_pétala'].std()

eixo.hist(df_is['comprimento_pétala'], bins = 20)
eixo.set_title('Iris Setosa', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da pétala', fontsize = 15)
eixo.grid(True)

eixo.annotate('$media = {0:.2f}$\n$desvio = {1:.2f}$' .format(media, desvio),
              xy = (1.7, 7), fontsize = 15)
eixo.axvline(media, color = 'black', linestyle = '--')
eixo.annotate('média', xy = (media - 0.2, 13.5), fontsize = 15)
eixo.axvline(df_is['comprimento_pétala'].median(), color = 'g', linestyle = '--')
eixo.annotate('mediana', xy = (df_is['comprimento_pétala'].median() + 0.05, 12), fontsize = 15,
              color = 'g')

fig.savefig('histograma_is.png', bbox_inches = 'tight')
```

```python
fig = plt.figure(figsize = (7,4))
eixo = fig.add_axes([0,0,1,1])

df_ivg = df[df['espécie'] == 'Iris-virginica']

media, desvio = df_ivg['comprimento_pétala'].mean(), df_ivg['comprimento_pétala'].std()

eixo.hist(df_ivg['comprimento_pétala'], bins = 20)
eixo.set_title('Iris Virginica', fontsize = 15, pad = 10)
eixo.set_xlabel('Comprimento da pétala', fontsize = 15)
eixo.grid(True)

eixo.annotate('$media = {0:.2f}$\n$desvio = {1:.2f}$' .format(media, desvio),
              xy = (6.3, 7), fontsize = 15)
eixo.axvline(media, color = 'black', linestyle = '--')
eixo.annotate('média', xy = (media + 0.1, 8), fontsize = 15)
eixo.axvline(df_ivg['comprimento_pétala'].median(), color = 'g', linestyle = '--')
eixo.annotate('mediana', xy = (df_ivg['comprimento_pétala'].median() + 0.1, 7), fontsize = 15,
              color = 'g')

fig.savefig('histograma_ivg.png', bbox_inches = 'tight')
```

```python
from PIL import Image

largura, altura = Image.open('histograma_ivg.png').size
combinada = Image.new('RGB', (3*largura, altura))
intervalo = 0
for imagem in map(Image.open, ['histograma_iv.png', 'histograma_is.png',
                               'histograma_ivg.png']):
  combinada.paste(imagem, (intervalo, 0))
  intervalo += largura

combinada.save('combinada.png')
```
