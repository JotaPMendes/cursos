```python
deaths = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv")
```

```python
def latest_by_country(data):
  return data.groupby('Country/Region').sum().iloc[:,-1]
```

```python
informations = [latest_by_country(df), latest_by_country(deaths), latest_by_country(recovered)]
combined = pd.concat(informations, axis = 1)
combined.columns = ['confirmed', 'deaths', 'recovered']
```

```python
sum_up = combined.sum()
letality_rate = sum_up['deaths'] / sum_up['confirmed']
letality_rate_2 = sum_up['deaths'] / (sum_up['deaths'] + sum_up['recovered'])
```

```python
letality_rate = combined['deaths'] / combined['confirmed'] * 100
letality_rate_2 = combined['deaths'] / (combined['deaths'] + combined['recovered']) * 100

combined['letality_rate'] = letality_rate
combined['letality_rate_2'] = letality_rate_2
combined.head()
```

```python
combined.sort_values('letality_rate', ascending = False)
```

```python
import seaborn as sns

sns.scatterplot(data = combined, x = 'confirmed', y = 'letality_rate')
```

```python
confirmed_40 = combined.query('confirmed > 40')
sns.distplot(combined.query('confirmed > 40')["letality_rate"], kde = False)
```
